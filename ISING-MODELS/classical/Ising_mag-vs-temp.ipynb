```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("mag_vs_temp.dat", skiprows=1)
T, M = data[:,0], data[:,1]

plt.plot(T, M, marker='o', linestyle='-')
plt.xlabel("Temperature (T)")
plt.ylabel("Magnetization per site")
plt.title("Magnetization vs Temperature")
plt.grid()
plt.show()
```
