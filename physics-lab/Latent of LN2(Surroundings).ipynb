```python
import scipy as sp
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()

#For P = 10 watt

time = np.arange(0,330,30)
timelr = np.arange(150, 330, 30)
masslr = [327.7, 327.7, 327.7, 327.7, 327.7, 327.7]
mass = [327.4, 327.4, 328.0, 328.0, 328.0, 327.7, 327.7, 327.7, 327.7, 327.7, 327.7]
x = timelr
y = masslr

res = stats.linregress(x, y)

plt.plot(time, mass, 'o')
plt.plot(x, y, 'o', label='original data')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.ylabel('Mass(g)')
plt.xlabel('Time(sec)')
plt.title('Power = 10 W')
plt.legend()
plt.show()

sp.stats.linregress(x, y)
```
