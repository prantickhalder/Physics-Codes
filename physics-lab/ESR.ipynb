```python
#Frequency vs Magnetic Field
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
rng = np.random.default_rng()

f = [15.1, 20.1, 25.1, 31.4, 36.1, 40, 45, 50, 55.1, 60, 65, 70, 75.7, 80.4, 85]
b = [0.5838, 0.7518, 0.9702, 1.2348, 1.3902, 1.5540, 1.7556, 1.9614, 2.1252, 2.3310, 2.5410, 2.7147, 2.9946, 3.1962, 3.3642]
b1 = np.zeros((len(b)))
f1 = np.zeros((len(b)))

for i in range(len(b)):
  f1[i] = 1000000*f[i]
  b1[i] = 0.001*b[i]

y = f1
x = b1

res = stats.linregress(x, y)
plt.plot(x, y, '.', label='original data')
plt.title('Frequency vs Magnetic Field')
plt.plot(x, res.slope*x, 'r', label='fitted line')
plt.ylabel('Frequency(Hz)')
plt.xlabel('Magnetic Field(T)')
plt.legend()
plt.show()

sp.stats.linregress(x, y)
```
