```python
#T vs t linear regression for the calculation of dQ(rod)/dt:

import pandas as pd
import scipy as sp
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()

df1 = pd.read_csv('/content/drive/MyDrive/ThermalCon/T vs t xls.CSV', encoding='unicode_escape')
df1.head()

time = df1['Time']
x = df1['Time']
y = df1['Run 1-Temperature(°C)']

res = stats.linregress(x, y)

plt.plot(x, y, '.', label='original data')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.xlabel('Time(sec)')
plt.ylabel('Temerature(°C)')
plt.legend()
plt.show()

sp.stats.linregress(x, y)
```
