```python
#Linear Regression for y=Bx

#Glycerine

import numpy as np
import matplotlib.pyplot as plt

x = [0.4405, 0.4500, 0.4300, 0.4310, 0.4425, 0.4395, 0.4415, 0.4465, 0.4515]
y = [17.7878, 18.1545, 17.3775, 17.4812, 17.7736, 17.6815, 17.8717, 18.0521, 18.0872 ]

sxy = 0
sx2 = 0
b = 0
sigy = 0
sigb = 0
sum = 0

n = len(x)

for i in range(n):
  sxy += x[i]*y[i]
  sx2 += x[i]**2

b = sxy/sx2

for i in range(n):
  sum += y[i] + b*x[i]

sigy = np.sqrt(sum/(n-1))

sigb = sigy/sx2


print('Slope, B =', b, 'MHz')
print('\nUncertainity in y=', sigy)
print('\nUncertainity in slope, b=', sigb, '*10^(-4)')

ycal = np.zeros((n))

for i in range(n):
  ycal[i] = b*x[i]

fig, ax = plt.subplots()
ax.plot(x, y, 'b*')
ax. plot(x, ycal, 'r')
plt.xlabel('Magnetic Field(T)')
plt.ylabel('Frequency(MHz)')
plt.title('Frequency v/s Magnetic Field for GLYCERINE', color='g')
plt.show()
```
