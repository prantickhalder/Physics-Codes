```python
#Calculations for Wavelenght vs 1/Frequency

import numpy as np

y = [0.1450, 0.1395, 0.1285, 0.1320, 0.0995]
x = [1/2400, 1/2500, 1/2700, 1/2900, 1/3500]

n = 5

ib = np.zeros((n))
i2 = np.zeros((n))
mi = np.zeros((n))
bmi = np.zeros((n))
sxy = 0
sx2 = 0
m = 0

#To calculate slope

for i in range(n):
  sxy += x[i]*y[i]
  sx2 += (x[i]**2)

m = sxy/sx2

print('The slope is:', m, '\n')

#Calculations

for i in range(n):
    ib[i] = y[i]*x[i]
    i2[i] = (x[i])**2
    mi[i] = m*x[i]
    bmi[i] = (y[i] - m*x[i])**2

for i in range(n):
    s = (y[i] - m*x[i])**2
    uncb = np.sqrt(((y[i] - m*x[i])**2)/(n - 1))

uncy = np.sqrt(s/(4))
uncm = uncb/np.sqrt(sx2)

print('The uncertainity in slope m is =', uncm, '\n')

print('sum y*x =', sxy,'\n', 'sum x^2 =', sx2, '\n')
print('y*x =', ib, '\n\n', 'x^2 =', i2, '\n\n', 'm*x =', mi,'\n\n', '(y - m*x)^2 =', bmi, '\n')
```

```python
#Standard Deviation

import numpy as np

#Frequency
n = [2400, 2500, 2700, 2900, 3500]
v = [350.25, 348.75, 346.95, 348.00, 348.25]
len = 5
s1 = 0
s2 = 0

for i  in range(len):
  s1 += v[i]**2
  s2 += v[i]

var = (s1/len) - (s2/len)**2
sd = np.sqrt(var)
vbest = s2/len
print('Standard deviation =', sd, '\nVariance = ', var)
print('v(best) =', vbest)

vel = vbest + sd/(np.sqrt(5))
print('Velocity =', vel)

```
