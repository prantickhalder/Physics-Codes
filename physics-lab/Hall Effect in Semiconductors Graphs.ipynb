```python
#Calibration of B with I

import numpy as np
import matplotlib.pyplot as plt

# n = 0

n = int(input("Enter the no. of data to input: "))
# input(n)

print(n)

cur = np.zeros((n))
mag = np.zeros((n))

print("\nEnter the values of current and applied magnetic field respectively:\n")

for i in range(n):
  cur[i] = float(input())

print('\n')

for i in range(n):
  mag[i] = float(input())

fig, ax = plt.subplots()
ax.plot(cur,mag, label='V vs I')
plt.legend()

#To plot V vs I

volt = np.zeros((n))
rvolt = np.zeros((n))
avgv = np.zeros((n))

print('\nEnter the values of forward and reverse voltage:\n')

for i in range(n):
  volt[i] = float(input())

print('\n')

for i in range(n):
  rvolt[i] = float(input())

print('\n')

for i in range(i):
  avgv[i]=(volt[i] + rvolt[i])/2

fig, ax = plt.subplots()
ax.plot(cur, avgv, 'r', label = 'Hall Voltage vs Current')
plt.legend()
```

```python
#Graph of Hall Voltage vs Current

import numpy as np
import matplotlib.pyplot as plt

cur = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0]
voltf = [0.4, 0.5, 0.6, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4]
voltr = [0.1, 0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1]

vh = np.zeros((10))

for i in range(10):
    vh[i] = (voltf[i] + voltr[i])/2
    print('Vall voltage with serial no.', (i+1), 'is', vh[i])

fig, ax = plt.subplots()
ax.plot(cur, vh, label = 'Hall Voltage vs Current')
plt.legend()
```
