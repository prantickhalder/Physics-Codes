```python
import pandas as pd
import scipy as sp
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()

df = pd.read_csv('/content/drive/MyDrive/ThermalCon/gradienttemp.CSV', encoding='unicode_escape')
df.head()

time = df['Time']
t2 = df['Lower Temp']
t1 = df['Upper Temp']
t0 = df['Cold Bath Temp']

#print(t0, t1, t2)

fig, ax = plt.subplots()
plt.xlabel('Time(sec)')
plt.ylabel('Temperature(°C)')
ax.plot(time, t1, 'b')
ax. plot(time, t2, 'r')


plt.grid(True)
plt.minorticks_on()
plt.grid(which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.show()

# t0 vs time linear regression:

lim = 450
x = time[lim:]
y = t0[lim:]

res = stats.linregress(x, y)
plt.plot(x, y, '.', label='original data')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.xlabel('Time(sec)')
plt.ylabel('Temperature(°C)')
plt.legend()
plt.show()

sp.stats.linregress(x, y)


```
