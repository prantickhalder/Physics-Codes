```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def func(m, T):
    return np.tanh(m / T) - m

def derivative(m, T):
    return (1 / (T * np.cosh(m / T)**2)) - 1

# Newton-Raphson Method
def newton_raphson(T, tol=1e-8, max_iter=100):
    m = 1.0  # Initial guess
    for _ in range(max_iter):
        f = func(m, T)
        df = derivative(m, T)
        if abs(f) < tol:
            return m
        if abs(df) < 1e-10:  # Avoid division by zero
            break
        m -= f / df
    return None  # Return None if not converged

# Generate T values and compute roots
T_values = np.arange(0.1, 1.2, 0.1)
roots = [newton_raphson(T) for T in T_values]

# Plot root vs. T
plt.figure(figsize=(8, 5))
plt.plot(T_values, roots, marker='o', linestyle='-')
plt.xlabel("T")
plt.ylabel("Root (m)")
plt.title("Root vs T (Newton-Raphson Method)")
plt.grid(True)
plt.show()
```
