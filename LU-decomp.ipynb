```python
#Q1. Write a python function that takes two matrices (A, B) of size n×n as an input and returns
#an output matrix C, where C = A × B. Test your function by taking two simple 3 × 3 matrices
#and compare your result with Numpy’s inbuilt function ”np.matmul(A,B)”.

import numpy as np
import sys

def mul(r1,c1,r2,c2,a,b):

  c = np.zeros((r1,c2))

  if(c1!=r2):
    sys.exit("c1 and r2 must be equal for matrix multiplication")

  for i in range(r1):
    for j in range(c1):
      a[i, j] = np.random.randint(1,10)

  for i in range(r2):
    for j in range(c2):
      b[i, j] = np.random.randint(1,10)

  for i in range(r1):
    for j in range(c2):
      for k in range(c1):
        c[i, j] += a[i, k]*b[k, j]

  return c

r1 = 6
c1 = 7
r2 = 7
c2 = 6

a = np.zeros((r1,c1))
b = np.zeros((r2,c2))

for i in range(r1):
  for j in range(c1):
    a[i, j] = np.random.randint(1,10)

for i in range(r2):
  for j in range(c2):
    b[i, j] = np.random.randint(1,10)

print('A is:\n', a,'\nB is:\n',b,'\n')
print('A*B is:\n', mul(r1,c1,r2,c2,a,b))

print('\nUsing matmul:\n', np.matmul(a,b),'\n\n')

```
