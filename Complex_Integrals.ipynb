```python
import numpy as np
from scipy.integrate import quad

# Define the complex function you want to integrate
def integrand(x):
    return np.exp(1j * x)  # Example of a complex function

# Integration limits
a = 0  # Lower limit
b = np.pi  # Upper limit

# Separate the real and imaginary parts of the integrand
real_part = lambda x: np.real(integrand(x))
imag_part = lambda x: np.imag(integrand(x))

# Perform the integration separately on the real and imaginary parts
real_integral, real_error = quad(real_part, a, b)
imag_integral, imag_error = quad(imag_part, a, b)

# Combine the results into a complex number
complex_integral = real_integral + 1j * imag_integral

print(f"Complex integral: {complex_integral}")
```
