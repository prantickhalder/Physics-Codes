```python
import numpy as np
import matplotlib.pyplot as plt

# Low pass

freq = [10, 20, 30, 40, 50, 100, 200, 300, 400, 500, 700, 1000, 1500, 2000, 3000, 4000, 5000, 8000, 9000, 10000, 12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 25000, 30000, 35000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 120000, 150000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000, 2000000, 3000000, 4000000, 5000000, 8000000, 10000000, 12000000, 15000000, 20000000]
vin = [1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.5, 1.13, 1.13, 1.11, 1.09, 1.07, 1.05, 1.02, 1.02, 1, .92, .92, .92, .92, 1.05, 1.15, 1.30, 1.35, 1.12]
vout = [12.5, 12.5, 12.3, 12.3, 12.3, 12.3, 12.3, 12.3, 12.3, 12.3, 12.3, 12.3, 12.3, 12.3, 12.3, 12.1, 12.1, 12.1, 12.1, 12.1, 12.1, 12.1, 11.9, 11.7, 11.7, 11.7, 11.7, 11.5, 11.3, 10.1, 8.8, 7.8, 7, 5.8, 5, 4.6, 4, 3.8, 3.4, 3, 2.61, 1.31, .9, .62, .5, .41, .346, .306, .269, .241, .117, .08, .07, .067, .062, .045, .0402, .0372, .0139]

n = len(freq)
gn = np.zeros(n)


for i in range(n):
  gn[i] = vout[i]/vin[i]

gain = 20*np.log10(gn)

logfreq = np.log10(freq)

fig, ax = plt.subplots()
ax.plot(logfreq, gain, label = '20log(Gain) vs log(Frequency)')
plt.xlabel('log(Frequency)')
plt.ylabel('20log(Gain)')
plt.title('Active Low Pass Filter')
plt.legend()

plt.grid(True)
plt.minorticks_on()
plt.grid(which='minor', color='#999999', linestyle='-', alpha=0.2)

g_max = np.max(gn)
g_rms = np.sqrt(np.mean(np.square(gn)))

plt.axhline(y=g_max*0.707, color='r', linestyle='-')
plt.axvline(x=5.1, color='g', linestyle='-')

print("Maximum value of g: ", g_max)
print("RMS value of g: ", g_rms)

plt.show()
```
