```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

freq = [20, 50, 100, 200, 500, 1000, 2000, 5000, 7000, 9000, 10000, 12000, 15000, 20000, 25000, 30000, 40000, 50000, 70000, 80000, 100000, 150000, 200000, 300000, 400000, 500000, 600000, 800000, 1000000, 1500000, 2000000, 3000000]
vin = [.115, .115, .115, .115, .117, .113, .115, .117, .115, .117, .115, .115, .115, .117, .117, .115, .115, .115, .115, .115, .115, .115, .115, .115, .115, .115, .115, .115, .115, .115, .115, .115]
vout = [.169, .314, .390, .470, .470, .440, .440, .45, .44, .44, .45, .45, .45, .46, .47, .48, .5, .52, .57, .59, .6, .59, .53, .43, .35, .28, .26, .193, .161, .113, .093, .064]

n1 = len(freq)
gn1 = np.zeros((n1))


#print(len(freq), len(vin), len(vout))

for i in range(n1):
  gn1[i] = vout[i]/vin[i]

#print(gn1)
gain1 = 20*np.log10(gn1)

print(gain1)

logfreq = np.log10(freq)

fig, ax = plt.subplots()
ax.plot(logfreq, gain1, label = '20log(Gain) vs log(Frequency)')
plt.xlabel('log(Frequency)')
plt.ylabel('20log(Gain)')
plt.title('Experimental Frequency Response Curve for MOSFET as an Amplifier')
plt.legend()

plt.grid(True)
plt.minorticks_on()
plt.grid(which='minor', color='#999999', linestyle='-', alpha=0.2)

g_max1 = np.max(gain1)
g_rms1 = np.sqrt(np.mean(np.square(gain1)))

plt.axhline(y=g_max1*0.707, color='r', linestyle='-')
#plt.axvline(x=4.56, color='g', linestyle='-')
#plt.axvline(x=5.86, color='g', linestyle='-')

print("Maximum value of g: ", g_max1)
print("RMS value of g: ", g_rms1)

plt.show()

print(len(freq),len(vin),len(vout),len(gain1))
data = pd.DataFrame({
    'Frequency (Hz)': freq,
    'Input Voltage (V)': vin,
    'Output Voltage (V)': vout,
    'Gain (dB)': gain1
})

# Print the DataFrame
print("Input Data:")
print(data)
```
