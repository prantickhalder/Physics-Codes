```python
import numpy as np
import matplotlib.pyplot as plt

n = 3
l = [30, 20, 10]
t0 = [1.94, 1.95, 1.95]
t1 = [1.57, 1.75, 1.88]
tb = [16.51, 36.19, 137.09]
tc = [1.74, 1.82, 1.92]

wbth = np.zeros((n))
wcth = np.zeros((n))
tbth = np.zeros((n))
tcth = np.zeros((n))
w0 = np.zeros((n))
w1 = np.zeros((n))
wb = np.zeros((n))
wc = np.zeros((n))
x = np.zeros((n))

for i in range(n):
  tbth[i] = (2*t0[i]*t1[i])/(t0[i] - t1[i])
  tcth[i] = (2*t0[i]*t1[i])/(t0[i] + t1[i])
for i in range(n):
  wbth[i] = 2*np.pi/tbth[i]
  wcth[i] = 2*np.pi/tcth[i]

for i in range(n):
  w0[i] = 2*np.pi/t0[i]
  w1[i] = 2*np.pi/t1[i]
  wb[i] = 2*np.pi/tb[i]
  wc[i] = 2*np.pi/tc[i]
  x[i] = (t0[i]**2 - t1[i]**2)/(t0[i]**2 + t1[i]**2)

fig, ax = plt.subplots()
ax.plot(l, w1, 'g')
plt.xlabel('Length(cm)')
plt.ylabel('W1 (1/s)')

fig, ax = plt.subplots()
ax.plot(l, wb, 'b')
plt.xlabel('Length(cm)')
plt.ylabel('Wb (1/s)')

fig, ax = plt.subplots()
ax.plot(l, wc, 'r')
plt.xlabel('Length(cm)')
plt.ylabel('Wc (1/s)')

fig, ax = plt.subplots()
ax.plot(l, x)
plt.xlabel('Length(cm)')
plt.ylabel('Degree of Coupling, X (1/s)')

for i in range(n):
  print('For length =' ,l[i],'cm', '\nPhase mode time period, T0 =', t0[i], '\nOut of phase time period, T1 =', t1[i], '\nTime period for beats,Tc =', tc[i], '\nTime period for beats, Tb =', tb[i] ,'\nBeat angular frequency, Wb =', wb[i], '\nCoupled oscillator angular frequency, Wc =', wc[i])
  print('Theoretical value of Tb =', tbth[i],'\nTheoretical value of Tc=', tcth[i], '\nTheoretical value of beat angular frequency, Wb =', wbth[i], '\nTheoretical value of coupled oscillator angular frequency, Wc =', wcth[i])
  print('Degree of coupling, X =',x[i],'\n' )
```
